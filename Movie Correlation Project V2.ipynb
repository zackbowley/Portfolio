```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots created


# Read in the data

df = pd.read_csv(r'C:\Users\zackb\OneDrive\Documents\movies.csv')


# Dropping the NaN values

df = df.dropna()


# Look at all the data

pd.set_option('display.max_rows' , None)
```

```python
# looking at the data

df.head()
```

```python
# Seeing if there is any missing data

for col in df.columns :
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data types for the columns

df.dtypes
```

```python
# Changing the data types of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
# Pulling the year from the released column
# Was previously missing before the data was updated

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df = df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
df.head()
```

```python
# Budget will have a high correlation with the gross
# Company will have a high correlation
# Directors will have a low correlation
```

```python
# Scatter plot with Budget vs. Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Film Budget')

plt.show()
```

```python
# Plot budget vs. gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
sns.regplot(x='votes', y='gross', data=df)
```

```python
# Taking a look at correlation

df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# There is High correlation between budget and gross
# The Speculation was right. The high the budget seems the be, the higher likely hood that it will earn more in gross
```

```python
# Creating a heatmap to better see the correlation between columns

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Cprrelation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
